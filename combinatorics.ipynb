Using the formula $C(n,r) = \frac{n!}{r!(n-1)!}$ to calculate the number of subsets of different sizes.

```python
def calculate_subset_counts(n):
    subset_counts = {}
    for k in range(n + 1):
        count = math.comb(n, k)
        subset_counts[k] = count
    return subset_counts
```

```python
subset_counts = calculate_subset_counts(30)
```

```python
sizes = list(subset_counts.keys())[:20] #First element is a 0 sized set
counts = list(subset_counts.values())[:20]

plt.figure(figsize=(12, 6))
plt.bar(sizes, counts)
plt.title(f'Number of Subsets by Size (n={30}) (Limitted to maximum subset size 19)')
plt.xlabel('Subset Size')
plt.ylabel('Number of Subsets')
plt.show()
```
